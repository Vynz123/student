````python


import tkinter as tk
from tkinter import messagebox
import random

# Define a class to represent a question
class Question:
    def __init__(self, question, options, correct_option):
        self.question = question
        self.options = options
        self.correct_option = correct_option

# Define the main quiz application class
class QuizApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Computer Science Quiz")
        self.root.configure(bg="#2c3e50")  # Set the background color of the window

        # List of questions with their options and correct answers
        self.questions = [
            Question("What does CPU stand for?", ["Central Processing Unit", "Computer Personal Unit", "Central Processor Unit", "Computer Processor Unit"], 0),
            Question("Which data structure uses LIFO order?", ["Queue", "Stack", "Array", "Linked List"], 1),
            # Add more questions here
        ]

        self.score = 0
        self.current_question_index = 0

        # Label to display the question
        self.question_label = tk.Label(root, text="", wraplength=600, bg="#2c3e50", fg="#000000", font=("Arial", 18, "bold"))
        self.question_label.pack(pady=20)

        # Create buttons for options
        self.option_buttons = []
        for i in range(4):
            button = tk.Button(root, text="", command=lambda i=i: self.check_answer(i), bg="#3498db", fg="#000000", font=("Arial", 14))
            button.pack(pady=5, padx=40, ipadx=10, ipady=5, fill=tk.X)  # Add padding and adjust button appearance
            self.option_buttons.append(button)

        self.next_question()

    def next_question(self):
        # Display the next question or show the score if all questions are done
        if self.current_question_index < len(self.questions):
            question = self.questions[self.current_question_index]
            self.question_label.config(text=question.question)

            for i, option in enumerate(question.options):
                self.option_buttons[i].config(text=option)

            self.current_question_index += 1
        else:
            self.show_score()

    def check_answer(self, selected_option):
        # Check the user's answer and update the score
        question = self.questions[self.current_question_index - 1]
        if selected_option == question.correct_option:
            self.score += 1

        # Move to the next question
        self.next_question()

    def show_score(self):
        # Display the final score in a messagebox and close the application
        messagebox.showinfo("Quiz Completed", f"Your score: {self.score}/{len(self.questions)}")
        self.root.destroy()

# Main function to create the GUI and run the application
def main():
    root = tk.Tk()  # Create the root window
    app = QuizApp(root)  # Create an instance of the QuizApp class
    root.mainloop()  # Run the main event loop

# Run the main function if this script is executed
if __name__ == "__main__":
    main()


````
